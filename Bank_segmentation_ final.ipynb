# Plotting Elbow Curve

Here we are plotting the elbow curve to extract the optimal cluster

```python
from sklearn.cluster import KMeans
Inter = []
for i in range(1,11):
  model1 = KMeans(n_clusters = i)
  model1.fit(x)
  Inter.append(model1.inertia_)
```

```python
Inter
```

```python
# plot with the help of dataframe
pd.DataFrame(Inter).plot()
```

Fitting the data into K_mean Clustering Algorithm

```python
# now by seeing the graph, we can put k = 3
new_model1 = KMeans(n_clusters = 3)
new_model1.fit(x)
```

```python
# how to see the clusters
clusters = new_model1.labels_
```

```python
print(clusters)
```

Let's Assign the Clusters

```python
X = x.assign(cluster = clusters)
```

```python
X['cluster'].value_counts()
```
